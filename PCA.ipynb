# 1. Dataset wine.csv

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
```

```python
wine= pd.read_csv('wine.csv')
wine
```

# Converting into numpy array

```python
df=wine.values
df
```

## Normalizing the numerical data

```python
df_norm=scale(df)
df_norm
```

```python
pca= PCA(n_components=14)
pca_values=pca.fit_transform(df_norm)
pca_values
```

```python
#loading or weights
pca.components_
```

```python
#the amount of variance that each PCA explain is
variance= pca.explained_variance_ratio_
variance
```

```python
# Cumulative variance
var1=np.cumsum(np.round(variance,decimals=6)*100)
var1
```

### Variance plot for PCA components obtained

```python
plt.plot(var1,color='green')
```

Using first 3 principal component scores

```python
pca_values[:,0:2]
```

```python
Wine_df=pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pca1','pca2','pca3'])],axis=1)
Wine_df
```

### Perform clustering using first 3 principal component scores

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=Wine_df,x='pca2',y='pca3',hue='pca1',s=100)
```

# Hierarchical cluctering

### Complete method of linkage

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
x=linkage(Wine_df,method='complete',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Hierarchical clustering dendrogram')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(x)
plt.show()
```

### Average method of linkage

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
x=linkage(Wine_df,method='average',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Hierarchical clustering dendrogram')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(x)
plt.show()
```

### Ward method of linkage

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
x=linkage(Wine_df,method='ward',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Hierarchical clustering dendrogram')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(x)
plt.show()
```
