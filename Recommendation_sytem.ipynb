# Problem statement.

### Build a recommender system by using cosine simillarties score.




```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
```

```python
!pip install chardet
import chardet
with open('/content/book (1).csv', 'rb') as f:
    encoding = chardet.detect(f.read())['encoding']
```

```python
book = pd.read_csv('/content/book (1).csv', encoding=encoding)
book
```

```python
book1=book.sort_values(by='User.ID').reset_index(drop=True)
book1
```

```python
book1.drop(columns='Unnamed: 0', axis=1)
```

## Number of unique users in the dataset

```python
len(book1['User.ID'].unique())
```

```python
len(book1['Book.Title'].unique())
```

```python
book1['User.ID'].duplicated().sum()
```

```python
book1.isnull().sum()
```

```python
book1 = book1.drop_duplicates(subset='User.ID', keep='last')
```

```python
book1.shape
```

```python
book1.drop(columns='Unnamed: 0', axis=1)
```

```python
user_book1=book1.pivot(index='User.ID', columns='Book.Title',
                       values='Book.Rating').reset_index()
```

```python
user_book1
```

```python
user_book1.index= book1['User.ID'].unique()
```

```python
user_book1.index.shape
```

## Replace those NaNs with the 0 value

```python
user_book1.fillna(0,inplace=True)
```

```python
user_book1.head()
```

## Calculating Cosine similarity between users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book1_co= 1-pairwise_distances(user_book1.values,metric='cosine')
```

```python
book1_co
```

### Store the  results in dataframe

```python
book1_df=pd.DataFrame(book1_co)
```

```python
book1_df
```

# Set the index and column name to userids

```python
book1_df.index=book1['User.ID'].unique()
book1_df.columns= book1['User.ID'].unique()
```

```python
book1_df.iloc[0:5]
```

```python
np.fill_diagonal(book1_co,0)
book1_df.iloc[0:5]
```

### fetching rating for Dead Sleep

```python
dead_rate= user_book1['Dead Sleep']
```

```python
pd.set_option('display.max_rows', None)
dead_rate[:50]
```

Finding the correlation with different books

```python
similar_to_Dead_Sleep= user_book1.corrwith(dead_rate)
```

```python
similar_to_Dead_Sleep.head(10)
```

```python
corr=pd.DataFrame(similar_to_Dead_Sleep,columns=['Correlation'])
corr.dropna(inplace=True)
corr.head()
```

```python
corr[corr['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)
```
